```python
## Loading the required libraries

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure
%matplotlib inline
import matplotlib.dates as mdates

plt.rcParams["figure.figsize"] = [12, 8]
plt.rcParams["figure.autolayout"] = True
```

```python
## Importing the dataset

Cor_Dataset=pd.read_csv(r'russia_losses_equipment.csv')
```

```python
## Fetching last 7 rows
Cor_Dataset.tail(7)
```

```python
## Finding descriptive statistics
Cor_Dataset.describe()
```

```python
## Let's find the missing data.i.e the null values

for col in Cor_Dataset.columns:
    pct_missing=np.mean(Cor_Dataset[col].isnull())
    print('{} - {}%'.format(col,pct_missing))
```

```python
## Finding mean value of column having null value  and replacing them with mean of it.
Mean_Special_Equipment=np.mean(Cor_Dataset['special equipment'])
Cor_Dataset['special equipment']=Cor_Dataset['special equipment'].fillna(Mean_Special_Equipment)
```

```python
Mean_Mobile_SRBM=np.mean(Cor_Dataset['mobile SRBM system'])
Cor_Dataset['mobile SRBM system']=Cor_Dataset['mobile SRBM system'].fillna(Mean_Mobile_SRBM)
```

```python
## Checking again if the null values are replaced with mean 
for col in Cor_Dataset.columns:
    pct_missing=np.mean(Cor_Dataset[col].isnull())
    print('{} - {}%'.format(col,pct_missing))
```

```python
Cor_Dataset
```

```python
## Finding the datatypes of colummns
Cor_Dataset.dtypes
```

```python
## converting the datatype from float to int
Cor_Dataset['special equipment']=Cor_Dataset['special equipment'].astype('int64')
Cor_Dataset['mobile SRBM system']=Cor_Dataset['mobile SRBM system'].astype('int64')
```

```python
## Checking if the datatypes is changed
Cor_Dataset.dtypes
```

```python
## Sorting values
Cor_Dataset.sort_values(by=['field artillery'],inplace=False,ascending=False)
```

```python
## Drop any duplicates

Cor_Dataset['date']=Cor_Dataset['date'].drop_duplicates().sort_values(ascending=False)
```

```python
## Scatterplot between date and field artillery
plt.scatter(Cor_Dataset['date'],Cor_Dataset['field artillery'],color='blue')
plt.xticks(np.arange(0, 38,step=3),rotation=90)
plt.xlabel('Date')
plt.ylabel('Field Artillery')
plt.title('Field Artillery losses in each date ')
plt.show()
```

```python
## Plot using Seaborn with Day vs Field artillery
sns.regplot(x='day',y='field artillery',data=Cor_Dataset,scatter_kws={'color':'blue'},line_kws={'color':'red'})
```

```python
## Correlation - methods - pearson, kendall, spearman
Cor_Dataset.corr()
```

```python
correlation_matrix=Cor_Dataset.corr(method='pearson')
sns.heatmap(correlation_matrix,annot=True)
plt.xlabel('Equipments')
plt.ylabel('Equipments')
plt.title('Correlation matrix for Losses of equipments')
plt.show()
```

```python
## Finding out high correlation
corr_pairs=correlation_matrix.unstack()
```
